<b> Now we query for the customer's Last Name, First Name, Track Names, and Album Titles.

We will only display the top 5 rows </b>

```python
sql_query = '''SELECT c.LastName, c.FirstName, t.Name, a.Title
        FROM Customer as c
        LEFT JOIN Invoice as i
        ON c.CustomerId = i.CuStomerId
        LEFT JOIN InvoiceLine as il
        ON i.InvoiceId = il.InvoiceId
        LEFT JOIN Track as t
        ON il.TrackID = t.TrackId
        LEFT JOIN Album as A
        ON t.AlbumID = a.AlbumId
        ORDER BY c.LastName, c.FirstName'''

pd.read_sql_query(sql_query, con = conn).head(5)
```
